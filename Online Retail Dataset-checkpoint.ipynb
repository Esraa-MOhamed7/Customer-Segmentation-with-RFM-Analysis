## RFM Analysis

```python
data["InvoiceDate"] = pd.to_datetime(data["InvoiceDate"])
```

```python
today_date = data["InvoiceDate"].max()
```

```python
rfm = data.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (today_date - x.max()).days,  # Recency
    "InvoiceNo": "nunique",                                # Frequency
    "TotalPrice": "sum"                                    # Monetary
})
```

```python
rfm
```

```python
rfm.columns = ["Recency", "Frequency", "Monetary"]
```

```python
rfm
```

```python
rfm=rfm[rfm["Monetary"]>0]
```

```python
rfm
```

```python
rfm.loc[:, "recency_score"] = pd.qcut(rfm["Recency"], 5, labels=[5, 4, 3, 2, 1])
rfm.loc[:, "frequency_score"] = pd.qcut(rfm["Frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
rfm.loc[:, "monetary_score"] = pd.qcut(rfm["Monetary"], 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm
```

```python
rfm.loc[:, "RFM_Score"] = (rfm["recency_score"].astype(str) + rfm["frequency_score"].astype(str))
```

```python
rfm.head()
```

```python
def segment_map(row):
    if row["recency_score"] == 5 and row["frequency_score"] == 5:
        return "Champions"
    elif row["recency_score"] >= 4 and row["frequency_score"] >= 4:
        return "Loyal Customers"
    elif row["recency_score"] == 5:
        return "Recent Customers"
    elif row["frequency_score"] == 5:
        return "Frequent Buyers"
    elif row["recency_score"] <= 2 and row["frequency_score"] <= 2:
        return "At Risk"
    else:
        return "Others"
```

```python
rfm.loc[:, "Segment"] = rfm.apply(segment_map, axis=1)
```

```python
rfm
```

## Visualizations

```python
plt.figure(figsize=(10,6))
sns.set_palette("husl")
sns.countplot(data=rfm, x="Segment", order=rfm["Segment"].value_counts().index)
plt.title("Customer Segment Distribution")
plt.xticks(rotation=45)
plt.ylabel("Number of Customers")
plt.xlabel("Segment")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))

# Pie 1: Top 3 by count
top_count = rfm["Segment"].value_counts().nlargest(3)
axes[0].pie(top_count.values, labels=top_count.index, autopct="%1.1f%%")
axes[0].set_title("Top 3 Segments by Count")

# Pie 2: Top 3 by revenue
top_revenue = rfm.groupby("Segment")["Monetary"].sum().nlargest(3)
axes[1].pie(top_revenue.values, labels=top_revenue.index, autopct="%1.1f%%")
axes[1].set_title("Top 3 Segments by Revenue")

plt.tight_layout()
plt.show()
```

```python
rfm_heatmap = rfm.groupby(["recency_score", "frequency_score"]).size().unstack()
sns.set_palette("husl")
sns.heatmap(rfm_heatmap,annot=True, fmt="d")
plt.title("Recency vs Frequency Heatmap")
plt.xlabel("Frequency Score")
plt.ylabel("Recency Score")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))
sns.set_palette("husl")
sns.histplot(rfm["Recency"], bins=30, ax=axes[0])
axes[0].set_title("Recency Distribution")
axes[0].set_xlabel("Days Since Last Purchase")
axes[0].set_ylabel("Number of Customers")

sns.histplot(rfm["Frequency"], bins=30, ax=axes[1])
axes[1].set_title("Frequency Distribution")
axes[1].set_xlabel("Number of Purchases")
axes[1].set_ylabel("Number of Customers")

plt.tight_layout()
plt.show()
```

```python
sns.set_palette("husl")
rfm.groupby("Segment")["Monetary"].sum().sort_values().plot(kind="barh")
plt.title("Total Revenue by Segment")
plt.xlabel("Total Monetary Value")
plt.ylabel("Segment")
plt.show()
```

```python
rfm["RFM_Score_Combo"] = (
    rfm["recency_score"].astype(str) +
    rfm["frequency_score"].astype(str) +
    rfm["monetary_score"].astype(str)
)

top_combos = rfm["RFM_Score_Combo"].value_counts().nlargest(10)
sns.set_palette("husl")
plt.figure(figsize=(10,5))
sns.barplot(x=top_combos.index, y=top_combos.values)
plt.title("Top 10 RFM Score Combinations")
plt.xlabel("RFM Score")
plt.ylabel("Customer Count")
plt.show()
```
